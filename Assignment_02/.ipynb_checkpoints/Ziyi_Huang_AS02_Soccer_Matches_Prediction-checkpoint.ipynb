```python
# Import libraries

#from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
import time
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Loading

```python
# Read the premier league dataset

df_premier_league = pd.read_csv("ft_df.csv")
df_premier_league
```

```python
#checking class proportions
winner_odds = (df_premier_league.ls_winner.value_counts(normalize = True) *100).round(2).reset_index()
winner_odds.columns = ['winner', '%']
winner_odds
```

```python
df_premier_league.info()
```

# Data Cleaning

### 1. Drop the columns that are not features or classes

```python
# drop columns: season, date, match_name, home_team, away_team, home_score, away_score, 
# h_match_points, a_match_points
list_drop_cols = ['season', 'date', 'match_name', 'home_team', 'away_team', 'home_score', 'away_score', 
                  'h_match_points', 'a_match_points']
df_premier_league.drop(columns = list_drop_cols, inplace = True)
df_premier_league.tail()
```

### 2. Fill in NaN cells with a specific value

```python
# As the dataset shows, there are NaN data cell due to the unavailable info of the original dataset
# so we can assign a specific value that makes it intepretable for ML models.
# Reference: https://towardsdatascience.com/whats-the-best-way-to-handle-nan-values-62d50f738fc
df_premier_league.fillna(value = -99, inplace = True)
df_premier_league.tail()
```

# Data Preparation

### 1. Store classes into an array y, assign classes into an integer value

```python
# Classes
# DRAW          : 0
# HOME_TEAM wins: 1
# AWAY_TEAM wins: 2

y = df_premier_league.iloc[:, 0].values
y = np.where(y == 'HOME_TEAM', 1, np.where(y == 'AWAY_TEAM', 2, 0))
y
```

### 2. Store features into an array X

```python
df_premier_league['ls_winner'] = np.where(df_premier_league.ls_winner == 'HOME_TEAM', 1, np.where(df_premier_league.ls_winner == 'AWAY_TEAM', 2, 0))
df_premier_league.head()
X = df_premier_league.iloc[:, df_premier_league.columns != 'winner'].values
X
```

### 3. Split the samples randomly into 80% training and 20% testing 

```python
#train_test_split function already shuffles the training sets internally before splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

### 4. Standardize the features using the MinMaxScaler class from scikit-learn

```python
# StandardScaler follows Standard Normal Distribution (SND). Therefore, it makes mean = 0 and scales the data to unit variance. 
# MinMaxScaler scales all the data features in the range [0, 1] or else in the range [-1, 1] if there are negative values in the dataset.
# Reference: https://www.geeksforgeeks.org/standardscaler-minmaxscaler-and-robustscaler-techniques-ml/
# Since my dataset contains negative values so I picked the MinMaxScaler

sc = MinMaxScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

# Machine Learning Modles Building

```python
# Logistic Regression

lr = LogisticRegression(max_iter = 1000, random_state=1) # parameter C is of overfitting and regularization
start = time.time()
lr_train_accuracy = lr.fit(X_train_std, y_train).score(X_train_std, y_train)
end = time.time()
print('Logistic Regression Training Accuracy:', '%.4f'%lr_train_accuracy)
lr_test_accuracy = lr.fit(X_train_std, y_train).score(X_test_std, y_test)
print('Logistic Regression Testing Accuracy:', '%.4f'%lr_test_accuracy)

lr_scores = cross_val_score(lr, X_train, y_train ,scoring= 'accuracy', cv=5)
print(" Clf result :", "%0.4f, +- %0.4f" % (lr_scores.mean(), lr_scores.std()))

lr_time = end-start
print('%.4f'%lr_time)
```

```python
tree = DecisionTreeClassifier(criterion='gini', max_depth=6, random_state=1)
start = time.time()
tree_train_accuracy = tree.fit(X_train_std, y_train).score(X_train_std, y_train)
end = time.time()
print('Decision Tree Training Accuracy:', '%.4f'%tree_train_accuracy)
tree_test_accuracy = tree.fit(X_train_std, y_train).score(X_test_std, y_test)
print('Decision Tree Testing Accuracy:', '%.4f'%tree_test_accuracy)

tree_scores = cross_val_score(tree, X_train, y_train ,scoring= 'accuracy', cv=5)
print(" Clf result :", "%0.4f, +- %0.4f" % (tree_scores.mean(), tree_scores.std()))

tree_time = end-start
print('%.4f'%tree_time)
```

```python
forest = RandomForestClassifier(max_samples = 200, random_state=1)
start = time.time()
forest_train_accuracy = forest.fit(X_train_std, y_train).score(X_train_std, y_train)
end = time.time()
print('Random Forest Training Accuracy:', '%.4f'%forest_train_accuracy)
forest_test_accuracy = forest.fit(X_train_std, y_train).score(X_test_std, y_test)
print('Random Forest Testing Accuracy:', '%.4f'%forest_test_accuracy)

forest_scores = cross_val_score(forest, X_train, y_train ,scoring= 'accuracy', cv=5)
print(" Clf result :", "%0.4f, +- %0.4f" % (forest_scores.mean(), forest_scores.std()))

forest_time = end-start
print('%.4f'%forest_time)
```

```python
# table1

lr_list = ["Logistic Regression", "%.4f"%lr_train_accuracy, "%0.4f +/- %0.4f" % (lr_scores.mean(), lr_scores.std()), 
           "%.4f"%lr_test_accuracy, "%.4f"%lr_time]

tree_list = ["Decision Tree", "%.4f"%tree_train_accuracy, "%0.4f +/- %0.4f" % (tree_scores.mean(), tree_scores.std()), 
           "%.4f"%tree_test_accuracy, "%.4f"%tree_time]

forest_list = ["Random Forest", "%.4f"%forest_train_accuracy, "%0.4f +/- %0.4f" % (forest_scores.mean(), forest_scores.std()), 
           "%.4f"%forest_test_accuracy, "%.4f"%forest_time]

pd.DataFrame([lr_list, tree_list, forest_list], columns=['Model', 'Training Accuracy', 'Cross Validation Score', 'Testing Accuracy', 'Training Time(sec)'])
```

# Filtering Feature Selection

```python
def condition_number(corr):# corr: correlation matrix
    eigValues,eigVectors = np.linalg.eig(corr) 
    return abs(max(eigValues)/min(eigValues))
 # greadily kick out the worst feature
def eliminate_multicol(X, threshold = 100):#weak corr threshold
    corr = np.corrcoef(X.T) # get corr matrix
    if (condition_number(corr) < threshold or X.shape[1]<=1): 
        return X  
    # find the feature with highest corr coef(lower triangle)
    mostCorrelated = np.argmax([max(corr[i+1:,i]) for i in range(X.shape[1]-1)] )
    choose = [True]*X.shape[1]
    choose[mostCorrelated] = False # kick that feature out
    return eliminate_multicol(X[:,choose])
```

```python
from sklearn.feature_selection import GenericUnivariateSelect as GUS, f_classif
X=eliminate_multicol(X) # remove multicollinearity
gus = GUS (f_classif, 'fwe', param=0.5) # keep features with P <0.5
X_new = gus.fit_transform(X, y) # remove irrelevant features
```
